```python
from flask import Flask, request
from pyngrok import ngrok

# Set your ngrok authtoken
ngrok.set_auth_token("2x2I0HFvCzMD63tMK1ZQgYiLwQI_RjgaeEzqxRcgQPcaQ6UD")

app = Flask(__name__)

@app.route("/incoming", methods=["POST"])
def incoming_message():
    msg = request.form.get('Body', '').strip().lower()
    sender = request.form.get('From')
    
    print(f"üì• Message from {sender}: {msg}")

    if msg == "on":
        print("üí° Simulated LED ON")
    elif msg == "off":
        print("üí° Simulated LED OFF")
    else:
        print("‚ö†Ô∏è Unknown command. Send 'on' or 'off'.")

    return "OK", 200

# Start ngrok tunnel
public_url = ngrok.connect(5000, bind_tls=True)
print("üîó Public URL for Twilio:", public_url)

# Run Flask app
app.run(port=5000)


```
