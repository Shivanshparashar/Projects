```python
import json
from sklearn.linear_model import LogisticRegression

#  Training data format 
with open("dataset.json", "r") as f:
    data = json.load(f)

X = [[entry["soil_moisture"], entry["threshold"]] for entry in data]
y = [1 if entry["soil_moisture"] < entry["threshold"] else 0 for entry in data]  # 1=ON, 0=OFF

model = LogisticRegression()
model.fit(X, y)

results = []
for entry in data:
    soil = entry["soil_moisture"]
    threshold = entry["threshold"]
    pred = model.predict([[soil, threshold]])[0]
    status = "ON" if pred == 1 else "OFF"
    results.append(f"Sensor data read: {soil}\nMotor Status: {status}")

with open("device_status.txt", "w") as f:
    f.write("\n".join(results))

print("Predictions saved to device_status.txt")
```
